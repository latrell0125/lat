### Outlier management

Manage the outlier in terms of 'OverallQual' and 'YearBuilt'

Choose any of the techniques for Outlier Management

```python
#Dropping the outlier rows with standard deviation

factor = 2
upper_limits = data['OverallQual'].mean() + data['OverallQual'].std () * factor
lower_limits = data['OverallQual'].mean() - data['OverallQual'].std () * factor
data = data[(data['OverallQual'] < upper_limits) & (data['OverallQual'] > lower_limits)]
data

factor = 2
upper_limits = data['YearBuilt'].mean() + data['YearBuilt'].std () * factor
lower_limits = data['YearBuilt'].mean() - data['YearBuilt'].std () * factor
data = data[(data['YearBuilt'] < upper_limits) & (data['YearBuilt'] > lower_limits)]
data
```

```python
#Dropping the outlier rows with Percentiles

upper_limits = data['OverallQual'].quantile(.99)
lower_limits = data['OverallQual'].quantile(.01)
data = data[(data['OverallQual'] < upper_limits) & (data['OverallQual'] > lower_limits)]
data

upper_limits = data['YearBuilt'].quantile(.99)
lower_limits = data['YearBuilt'].quantile(.01)
data = data[(data['YearBuilt'] < upper_limits) & (data['YearBuilt'] > lower_limits)]
data
```

```python
#Capping the outlier rows with percentiles

upper_limits = data['OverallQual'].quantile(.99)
lower_limits = data['OverallQual'].quantile(.01)

upper_limits = data['YearBuilt'].quantile(.99)
lower_limits = data['YearBuilt'].quantile(.01)
```

```python
data.loc[(data['OverallQual'] > upper_limits), 'OverallQual'] = upper_limits
data.loc[(data['OverallQual'] < lower_limits), 'OverallQual'] = lower_limits
data

data.loc[(data['YearBuilt'] > upper_limits), 'YearBuilt'] = upper_limits
data.loc[(data['YearBuilt'] < lower_limits), 'YearBuilt'] = lower_limits
data
```

### Balancing the dataset

Balanced the dataset in terms of "SaleCondition" (Normal, Abnormal and Partial)

```python
#insert code here
plt.figure(figsize=(8,6))

data.SaleCondition.value_counts().plot(kind='bar')

plt.ylabel('Number of data points')
plt.xlabel('Sale Condition')
plt.show()
```

```python
countby = data.SaleCondition.value_counts(dropna=False,)
countby
```

```python
def sampling_k_elements(SaleCondition, k=86):
    if len(SaleCondition) < k:
        return SaleCondition
    return SaleCondition.sample(k)

balanced = data.groupby('SaleCondition').apply(sampling_k_elements).reset_index(drop=True)
```

```python
count = balanced.SaleCondition.value_counts(dropna=False)
count
```

```python
plt.figure(figsize=(10,5))
plt.pie(count, labels=count.index)
plt.title('Pie chart showing balanced distibution of values for Sale Condition Categories')
plt.show()
```
